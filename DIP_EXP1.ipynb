### (3)	Convert Color Image to Grayscale. 

(a) Display the original Lena image and the grayscale images obtained by three grayscaling methods in the same figure. Add the corresponding title. 1)Maximum of the three components; 2) The average of the three components; 3) gray=0.30*r + 0.59*g + 0.11*b.

(b) Answer the question: What are their differences?

```python
#load the image and display the original Lena image
Lena=Image.open('../Exp1/images/lena.jpg')
Lena=np.array(Lena)
plt.subplots_adjust(wspace=0,hspace=1)
subplot=plt.subplot(2,2,1)
plt.imshow(Lena)
subplot.set_title('original')
#print(Lena.dtype)

#Convert original image to grayscale with three methods
#Maximum of the three components
#get the row and column of the image
row,column=Lena.shape[0:2]
#print(row,column)
#set an image to store the grayscale
max_gray=np.zeros((row,column),dtype=Lena.dtype)
#set the maximun of three components into the grayscale
for i in range(row):
    for j in range(column):
        max_gray[i,j]=max(Lena[i,j,0],Lena[i,j,1],Lena[i,j,2])
#we can also use another method: np.max(Lena,axis=2)
#max_gray=np.max(Lena,axis=2)

#display the image using the maximum method
subplot=plt.subplot(2,2,2)
plt.imshow(max_gray)
subplot.set_title('Maximum')

#The average of the three components
#Because when we add the three components the result will overflow, so we should change the type
average_gray=np.zeros((row,column),dtype=float)
#A simple method to change the image into float type is: float_Lena=Lena/255
#Another useful method is using skimage package
float_Lena=ski.img_as_float(Lena)
#print(float_Lena)
for i in range(row):
    for j in range(column):
        average_gray[i,j]=(float_Lena[i,j,0]+float_Lena[i,j,1]+float_Lena[i,j,2])/3

average_gray=ski.img_as_ubyte(average_gray)
#to get the average of components we can also use another method: np.mean(Lena,axis=2)
#average_gray=np.mean(Lena,axis=2)
#print(average_gray)

#display the image using the average method
subplot=plt.subplot(2,2,3)
plt.imshow(average_gray)
subplot.set_title('Average')

#gray=0.30*r+0.59*g+0.11*b
weight_gray=np.zeros((row,column),dtype=Lena.dtype)
for i in range(row):
    for j in range(column):
        weight_gray[i,j]=int(0.30*Lena[i,j,0]+0.59*Lena[i,j,1]+0.11*Lena[i,j,2])

#display the image using the weightedaverage methods
subplot=plt.subplot(2,2,4)
plt.imshow(weight_gray)
subplot.set_title('Weighted verage')

#Answer the question: What are their differences?
#1.The maximum method is to find the maximum value among the three-channel components of the color image, and then use the obtained maximum value as the grayscale value of the grayscale image.
#2.The average method is to take the average of the three-channel components of the color image, and then use the average value as the gray value of the grayscale image.
#3.The weighted average method is to weight and average the three-channel components of the color image according to the importance of each channel, and then use the obtained weighted average as the gray value of the grayscale image.
#4.The maximum method makes the image brighter, the average method makes the image smoother, but some detailed features are lost, and the weighted average method highlights the detailed features.
```

### (6)	Image Denoising by Averaging. 

(a) Load an RGB image. 

(b) Add Gaussian noise with a mean value of 0 and a variance of 0.1 to it. 

(c) Display and compare the images before and after adding noise. 

(d) Use the ``for`` loop to add 3, 30, and 300 images with random Gaussian noise and find their average value, respectively. 

(e) Display these averaged images in the same figure. Add corresponding titles, and compare the results.

```python
#load an RGB image
lena_path='../Exp1/images/lena.jpg'
lena=io.imread(lena_path)

#add gaussian noise with a mean value of 0 and a variance of 0.1 to it
gaussian_noise_img=ski.util.random_noise(lena,mode='gaussian',mean=0,var=0.1)

#display and compare the images before and after adding noise
plt.subplots_adjust(wspace=1,hspace=1)
subplot=plt.subplot(2,3,1)
plt.imshow(lena)
subplot.set_title('original')
subplot=plt.subplot(2,3,2)
plt.imshow(gaussian_noise_img)
subplot.set_title('with_gaussian_noise')

compare_block=np.zeros((lena.shape),dtype=lena.dtype)
compare_block=(lena==gaussian_noise_img)
print(compare_block)

#use the for loop to add 3 images with random Gaussian noise and find their average value,respectively
num_images = [3,30,300]
def process(num_images):
    noisy_images=[]
    for i in range(num_images):
        lena = Image.open(lena_path)
        noisy_image = ski.util.random_noise(np.array(lena), mode='gaussian', var=0.1)
        noisy_images.append(noisy_image)
    #find its average value
    average_image = np.mean(np.array(noisy_images), axis=0)
    #Image.fromarray((average_image * 255).astype(np.uint8))
    return average_image
#display this average image and add title
title=['3_images','30_images','300_images']
average_images=[]
for i in range(3):
    subplot=plt.subplot(2,3,4+i)
    subplot.set_title(title[i])
    average_image=process(num_images[i])
    plt.imshow(average_image)
    average_images.append(average_image)

#compare the result
for i in range(2):
    print('round:',i+1)
    compare_block=((average_image[i+1]-lena)<(average_image[i]-lena))
    print(compare_block)
```

### (9)	Adding image mask. 

(a) Load the Lena image with OpenCV. 

(b) Adding a circular mask on the Lena image. (<font style="color:blue;">Tips: You may use``numpy.ogrid``</font>). An example of the generated result is shown in the figure below.
 
![lena.jpg](./images/exp1_9.png) 
<center><font style="color:red;">** Figure 2. The result of adding binary mask to Lean image** </font></center>

```python
#load the Lena image with OpenCV
lena=cv2.imread('../Exp1/images/lena.jpg')

#create the circle mask
row,column=lena.shape[0:2]
radius = min(row,column)//2
y,x = np.ogrid[:row,:column]
center_x=column // 2
center_y=row // 2
mask = ((x - center_x) ** 2 + (y - center_y) ** 2) <= radius ** 2

# Apply mask to Lena image
lena[~mask] = 0

#change the mode of image and display it
lena=cv2.cvtColor(lena,cv2.COLOR_BGR2RGB)
plt.imshow(lena)
```

### (10)	Capture Images by Laptop Camera with OpenCV. (``Bonus Task``)

(a) Read the video stream from the laptop camera.(For more details check [this](http://www.1zlab.com/wiki/python-opencv-tutorial/opencv-read-usb-camera-image/)) 

(b) Implement Time-lapse photography.(It can be used to record the whole process of the evaporation of water droplets in the cup. <font style="color:blue;"> Tips: You may use``time.sleep``</font>)

(c) Generate corresponding MP4 video or GIF image.

```python
#define a function to read the video stream from the laptop camera
def read_video_stream():
    #get default laptop camera
    cap = cv2.VideoCapture(0)

    #judge whether the capture is open
    while(cap.isOpened()):
        #get the frame from the camera
        ret, frame = cap.read()

        if ret:
            #display the frame
            cv2.imshow('frame', frame)

            # break when receiving 'q'
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    # release the camera
    cap.release()
    # distroy all window
    cv2.destroyAllWindows()

read_video_stream()

#define a function to implement Time-lapse photography
def Time_lapse_photography(file_prefix,duration,interval):
    #get default laptop camera
    cap = cv2.VideoCapture(0)
    #initialize frame counter
    frame_count=0
    
    #capture frames for the specified duration
    start_time=time.time()
    while time.time()-start_time<duration:
        ret, frame = cap.read()
        if ret:
            # Write the frame into the video at the specified interval
            if frame_count%interval==0:
                filename = f"{file_prefix}_{frame_count}.jpg"
                cv2.imwrite(filename,frame)
                print(f'Saved frame {frame_count} as {filename}')
            frame_count+=1
            #Wait for 1 second between captures
            time.sleep(1)
        else:
            break
    #release the camera
    cap.release()
#capture frames every 5 seconds for 1 minute
Time_lapse_photography('time_lapse',60,5)

#define a function to generate MP4 video
def generate_video(file_prefix, duration, interval, output_file):
    # Get default laptop camera
    cap = cv2.VideoCapture(0)
    
    # Define the codec and create VideoWriter object
    # Codec for MP4 video
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
    # 20 fps, resolution (640x480)
    out = cv2.VideoWriter(output_file, fourcc, 20.0, (640, 480))  
    
    # Initialize frame counter
    frame_count = 0
    
    # Capture frames for the specified duration
    start_time = time.time()
    while time.time() - start_time < duration:
        ret, frame = cap.read()
        if ret:
            # Write the frame into the video at the specified interval
            if frame_count % interval == 0:
                out.write(frame)
                print(f"Added frame {frame_count} to the video")
            frame_count += 1
            # Wait for 1 second between captures
            time.sleep(1)  
        else:
            break
    
    # Release the camera and video writer
    cap.release()
    out.release()
    cv2.destroyAllWindows()

generate_video('time_lapse', 200, 5, 'output_video.mp4')

#define a function to generate gif images
def generate_gif(file_prefix, duration, interval, output_file):
    images = []
    # Capture frames for the specified duration
    for i in range(0, duration, interval):
        image_path = f"{file_prefix}_{i}.jpg"
        if os.path.exists(image_path):
            images.append(imageio.v2.imread(image_path))
        else:
            print(f"Image {image_path} not found.")
    
    # Write the images to a GIF file
    if images:
        imageio.mimsave(output_file, images, duration=1)
        print(f"Generated GIF: {output_file}")
    else:
        print("No images found.")

generate_gif('time_lapse', 60, 5, 'output.gif')
```
